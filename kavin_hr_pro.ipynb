HR PROJECT

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/hr dataset.csv')
df.head()
```

```python
# checking info
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1, inplace=True, errors='ignore')
df.shape
```

```python
df.describe()
```

```python
sns.countplot(x='Attrition', data=df)
plt.show()
```

separating categorical and numerical columns

```python
cat_cols = []
num_cols = []

for col in df.columns:
    if df[col].dtype == 'object':
        cat_cols.append(col)
    else:
        num_cols.append(col)

print(cat_cols)
print(num_cols)
```

Label Encoding to convert text to numbers

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])

df.head()
```

checking correlation

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), annot=True, fmt='.1f')
plt.show()
```

spliting data into training and testing

```python
# separating x and y
x = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
# model build
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(x_train, y_train)
pred_lr = lr.predict(x_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred_lr))
print(confusion_matrix(y_test, pred_lr))
```

```python
# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
pred_dt = dt.predict(x_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, pred_dt))
print(confusion_matrix(y_test, pred_dt))
```

```python
# Random Forest
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
pred_rf = rf.predict(x_test)
print("Random Forest Accuracy:", accuracy_score(y_test, pred_rf))
print(confusion_matrix(y_test, pred_rf))
```

```python
# KNN
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
pred_knn = knn.predict(x_test)
print("KNN Accuracy:", accuracy_score(y_test, pred_knn))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(rf, x, y, cv=5)
print(score)
print("Mean CV Score:", score.mean())
```

```python
# saving the best model
import joblib
joblib.dump(rf, 'HR_Analytics_Model.pkl')
```

Trying The Prediction

```python
print("Actual value:", y_test.iloc[5])
print("Predicted value:", rf.predict(x_test.iloc[5].values.reshape(1, -1)))
```
